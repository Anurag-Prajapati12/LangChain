```python
import os
from langchain_core.tools import tool
from langchain_google_genai import ChatGoogleGenerativeAI
import requests
from langchain_core.messages import HumanMessage
```

```python
@tool
def multiply(a: int,b: int) -> int:
    """Given 2 number a and b this tool returns thier product"""
    return a * b
```

```python
print(multiply.invoke({"a": 5, "b": 6}))  # should return 30
```

```python
llm = ChatGoogleGenerativeAI(model="models/gemini-2.5-flash", temperature=0)
```

```python
llm_with_tools = llm.bind_tools([multiply]) #tool binding
```

```python
llm_with_tools.invoke('hi how are you')
```

```python
result = llm_with_tools.invoke('multiply 3 with 10')
```

```python
result.tool_calls[0]['args']
```

```python
multiply.invoke(result.tool_calls[0])
```

```python
query = HumanMessage('multiply 3 with 10')
```

```python
messages = [query]
```

```python
result = llm_with_tools.invoke(messages)
```
